```python
from stiefelpol import Stiefel
import convexPol
```

### Generation of a Random Polygon from a Point in $\text{St}_2(\mathbb R^{10})$  ###

First, we will sample a random point from $\text{St}_2\left(\mathbb R^{10}\right)$. In the example below, we will use the function `StiefelSample` to generate a $10 \times 2$ Gaussian matrix and perform a $QR$ decomposition.

```python
frame = Stiefel.StiefelSample(10)
```

Now, let's map this matrix to a polygon with 10 edges.

```python
edges = Stiefel.FrameToEdges(frame)

Stiefel.viewPolygon(Stiefel.EdgesToVertices(edges))
```

### Transforming a Random Polygon into a Convex Polygon ###

```python
edges_plane_ordenado = convexPol.orderAngle(edges)

vertices_ordenado = Stiefel.EdgesToVertices(edges_plane_ordenado)

Stiefel.viewPolygon(vertices_ordenado)
```
